```python
#Q7
isolado=int(input("Informe o valor do vértice isolado: "))
l2=int(input("Informe o valor do segundo vértice: "))
l3=int(input("Informe o valor do terceiro vértice: "))
if l2+l3 > isolado:
  print("os valores podem formar um triangulo")
  if l2==l3 and l3 != isolado:
    print("o qual é um isosceles")
  elif l2 == l3 and l3 == isolado:
    print("o qual é um Equilatero")
  elif l2 != l3 and l3 != isolado:
    print("o qual é um Escaleno")
else:
  print("Os valores não forman um triangulo")
```

```python
#Q8
print("qual a forma geometrica você quer saber a area?")
forma=int(input("1-Triangulo, 2-Retangulo, 3-Circulo "))
if forma ==1:
  print("Triangulo")
  base=float(input("digite a Base "))
  altura=float(input("digite a altura "))
  result=(base*altura)/2
  print(f"A area do seu triangulo é:{result:,.2f}")

elif forma ==2:
  print("Retangulo")
  base=float(input("digite a Base "))
  altura=float(input("digite a altura "))
  result=base*altura
  print(f"A area do seu Retangulo é:{result:,.2f}")

elif forma ==3:
  print("Circulo")
  raio=float(input("digite o raio "))
  result=3.14*(raio ** 2)
  print(f"A area do seu circulo:{result:,.2f}")
```
